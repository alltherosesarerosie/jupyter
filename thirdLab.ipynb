```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

# 2.	Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
x = np.linspace(-1, 1, 10)
y = 2*x ** 2 - x - 2
plt.title('Quadratic function', fontsize='x-large', color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot(x, y, 'm')
plt.show()
```

# 3.	Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
x = np.linspace(-1 * np.pi, 1 * np.pi, 1000)
y = np.sin(x) + 5.5 * np.cos(2 * x)
plt.title('Harmonic function', fontsize='x-large', color='b')
plt.plot(x, y, 'm--')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

# 4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
#first
x = np.linspace(-1, 1, 10)
y1 = 2*x ** 2 - x - 2
plt.subplot(1, 2, 1)
plt.title('Quadratic function', fontsize='x-large', color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot(x, y1)

#second
x = np.linspace(-1 * np.pi, 1 * np.pi, 1000)
y2 = np.sin(x) + 5.5 * np.cos(2 * x)
plt.subplot(1, 2, 2)
plt.title('Harmonic function', fontsize='x-large', color='b')
plt.plot(x, y2, 'm--')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

# 5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

```python
x1 = np.linspace(-1, 1, 100)
y1 = 2 * x1**2 - x1 - 2

x2 = np.linspace(-1 * np.pi, 1 * np.pi, 1000)
y2 = np.sin(x2) + 5.5 * np.cos(2 * x2)

plt.figure(figsize=(5, 5))  

plt.plot(x1, y1, 'r:', label='Quadratic function: $2x^2 - x - 2$')
plt.plot(x2, y2, 'm--', label='Harmonic function: $sin(x) + 5.5cos(2x)$')

plt.xlabel('x')
plt.ylabel('y')

plt.title('Graph of Quadratic and Harmonic function', fontsize='x-large', color='b')

plt.legend()
plt.grid(True)
plt.show()
```

# 	6. Построить график функции, заданной параметрически,  x(t)=cos⁡〖(3t),y=sin⁡〖(2t),t∈[0,4π]〗 〗


```python
t = np.linspace(0, 4 * np.pi, 1000)

# Задаем параметрические уравнения x(t) и y(t)
x = np.cos(3 * t)
y = np.sin(2 * t)

# Создаем график
plt.figure(figsize=(10, 10))
plt.plot(x, y, label='Parametric function: $x(t) = cos(3t), y(t) = sin(2t)$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parametric function', fontsize='x-large', color='b')
plt.legend()
plt.grid(True)
plt.show()
```
